Práctica 1: Introducción al Procesamiento de Imágenes
Dibujar con Python y OpenCV la primera letra del nombre en un lienzo

```python
import cv2
import numpy as np
```

```python
# Creamos el lienzo en las dimenciones de 400x400 pixeles
lienzo = np.ones((alto, ancho, 3), dtype=np.uint8) * 255
```

```python
# Definimos los colores
color_letra = (200, 100, 0)
color_borde = (50, 50, 50)
```

```python
# Dibujamos el marco
cv2.rectangle(lienzo, (10, 10), (ancho - 10, alto - 10), color_borde, 2)
```

```python
# Dibujamos la letra sobre el lienzo
cv2.putText(lienzo, letra, (x, y), fuente, escala, color_letra, grosor, cv2.LINE_AA)
```

```python
# Calculamos el tamaño del texto para centrarlo en el lienzo
(ancho_texto, alto_texto), _ = cv2.getTextSize(letra, fuente, escala, grosor)
x = (ancho - ancho_texto) // 2
y = (alto + alto_texto) // 2
```

```python
# Guardar el resultado
cv2.imwrite("letra_S.png", lienzo)
```
